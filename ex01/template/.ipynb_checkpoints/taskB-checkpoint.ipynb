```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
```

Data Generation
===

```python
np.random.seed(10)
p, q = (np.random.rand(i, 2) for i in (4, 5))
p_big, q_big = (np.random.rand(i, 80) for i in (100, 120))

print(p, "\n\n", q)
```

Solution
===

```python
def naive(p, q):
    ''' fill your code in here...
    '''
    'p,q are px2 and qx2 matrices'
    
    'Number of points in p'
    numP,_=np.shape(p)
    
    'Number of points in q'
    numQ,_=np.shape(q)
    
    'used for storing distance of each point'
    D=np.zeros((numP,numQ))

    for i in range(0,numP):
        for j in range(0,numQ):
            D[i][j]=np.sqrt(np.power((p[i][0]-q[j][0]),2)+np.power((p[i][1]-q[j][1]),2))
    
naive(p,q)    
```

### Use matching indices

Instead of iterating through indices, one can use them directly to parallelize the operations with Numpy.

```python
rows, cols = np.indices((p.shape[0], q.shape[0]))
print(rows, end='\n\n')
print(cols)
```

```python
print(p[rows.ravel()], end='\n\n')
print(q[cols.ravel()])T
```

```python
def with_indices(p, q):    
    return np.sqrt(np.sum(np.power(p[rows.ravel()]-q[cols.ravel()],2),1))
```

### Use a library

`scipy` is the equivalent of matlab toolboxes and have a lot to offer. Actually the pairwise computation is part of the library through the `spatial` module.

```python
from scipy.spatial.distance import cdist

def scipy_version(p, q):
    return cdist(p, q)
```

### Numpy Magic

```python
def tensor_broadcasting(p, q):
    return np.sqrt(np.sum((p[:,np.newaxis,:]-q[np.newaxis,:,:])**2, axis=2))
```

# Compare methods

```python
methods = [naive, with_indices, scipy_version, tensor_broadcasting]
timers = []
for f in methods:
    r = %timeit -o f(p_big, q_big)
    timers.append(r)
```

```python
plt.figure(figsize=(10,6))
plt.bar(np.arange(len(methods)), [r.best*1000 for r in timers], log=False)  # Set log to True for logarithmic scale
plt.xticks(np.arange(len(methods))+0.2, [f.__name__ for f in methods], rotation=30)
plt.xlabel('Method')
plt.ylabel('Time (ms)')
plt.show()
```
