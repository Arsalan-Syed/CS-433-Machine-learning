# 2 Grid Search

Fill in the function `grid_search()` below:

```python
def grid_search(y, tx, w0, w1):
    """Algorithm for grid search."""
    losses = np.zeros((len(w0), len(w1)))
    
    # ***************************************************
    # INSERT YOUR CODE HERE
    # TODO: compute loss for each combination of w0 and w1.
    # ***************************************************
    for a in range (0,len(w0)):
        for b in range(0,len(w1)):
            losses[a][b]=compute_loss(y, tx, [w0[a],w1[b]])
    return losses

grid_search(y,tx,[0,1,2],[0,1,2])
```

# 3 Gradient Descent

Again, please fill in the functions `compute_gradient` below:

```python
def compute_gradient(y, tx, w):
    """Compute the gradient."""
    # ***************************************************
    # INSERT YOUR CODE HERE
    # TODO: compute gradient and loss
    # ***************************************************
    '''e=y-X*w'''
    N=y.shape[0]
    e=y-np.dot(tx,w)
    return -1/N*np.dot(np.transpose(tx),e)
compute_gradient(y,tx,[1,2])
```

Please fill in the functions `gradient_descent` below:

```python
def gradient_descent(y, tx, initial_w, max_iters, gamma):
    """Gradient descent algorithm."""
    # Define parameters to store w and loss
    ws = [initial_w] 
    losses = []
    w = initial_w
    for n_iter in range(max_iters):
        # ***************************************************
        # INSERT YOUR CODE HERE
        # TODO: compute gradient and loss
        # ***************************************************
        loss=compute_loss(y, tx, w)
        grad=compute_gradient(y, tx, w)
        'raise NotImplementedError'
        # ***************************************************
        # INSERT YOUR CODE HERE
        # TODO: update w by gradient
        # ***************************************************
        w=w-gamma*grad
        'raise NotImplementedError'
        # store w and loss
        ws.append(w)
        losses.append(loss)
        print("Gradient Descent({bi}/{ti}): loss={l}, w0={w0}, w1={w1}".format(
              bi=n_iter, ti=max_iters - 1, l=loss, w0=w[0], w1=w[1]))

    return losses, ws
```

# 4 Stochastic gradient descent

```python
def compute_stoch_gradient(y, tx, w):
    """Compute a stochastic gradient from just few examples n and their corresponding y_n labels."""
    # ***************************************************
    # INSERT YOUR CODE HERE
    # TODO: implement stochastic gradient computation.It's same as the gradient descent.
    # ***************************************************
    '''
    batch_size=10
   
    sumLoss=0
    for i in range(0,batch_size):
        sumLoss=sumLoss+compute_loss(y,tx,w)
    sumLoss=1/N*sumLoss
    '''
        
    N=y.shape[0]
    e=y-np.dot(tx,w)
    return -1/N*np.dot(np.transpose(tx),e)
    

def stochastic_gradient_descent(y, tx, initial_w, batch_size, max_iters, gamma):
    """Stochastic gradient descent algorithm."""
    # ***************************************************
    # INSERT YOUR CODE HERE
    # TODO: implement stochastic gradient descent.
    # ***************************************************
    samples=batch_iter(y, tx, batch_size, num_batches=1, shuffle=True)
    ws=[initial_w]
    losses=[]
    w=initial_w
   
    '''have to use samples, not entire input'''
    for n_iter in range(max_iters):
        sumGrad=0
        loss=0
        for minibatch_y, minibatch_tx in batch_iter(y, tx,batch_size):
            loss=loss+compute_loss(minibatch_y, minibatch_tx, w)
            sumGrad=sumGrad+compute_stoch_gradient(minibatch_y,minibatch_tx, w)
        g=sumGrad/batch_size
        
        w=w-np.multiply(gamma,g)
        
        ws.append(w)
        losses.append(loss)
    
    return losses, ws

stochastic_gradient_descent(y, tx,[1,2], 10, 1, 0.1)
```
